```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define the range for the x-axis
x = np.linspace(-4, 4, 1000)

# Define the normal distribution
mean = 0
std_dev = 1
pdf = norm.pdf(x, mean, std_dev)

# Create a figure and adjust the figure size
plt.figure(figsize=(12, 6))

# Plot the normal distribution curve
plt.plot(x, pdf, label='Normal Distribution', color='black', linewidth=4)

# Highlight the area under the curve where z-score >= 3.0
x_fill = np.linspace(0, 2.5, 1000)
y_fill = norm.pdf(x_fill, mean, std_dev)
plt.fill_between(x_fill, y_fill, color='red', alpha=0.5, label='Z-score >= 3.0')

# Add a vertical line in the middle
plt.axvline(x=0, color='blue', linestyle='--', linewidth=2, label='Mean (Z=0)')

# Remove the axes
plt.axis('off')

# Save the plot as a transparent PNG
plt.savefig('normal_distribution_2.png', transparent=True, bbox_inches='tight', pad_inches=0)

# Show the plot (optional)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
model_names = ['Adult', 'California', 'Abalone','Diabetes']
scores = {
        "Non-Watermark": [.835, .826, .847, .831],
        "T-REST": [.779, .808, .779, .787],
        "EXP": [.37, .364, 0, .333]
    }
            
print(scores)

x = np.arange(len(model_names))  # the label locations
width = 0.25  # the width of the bars
multiplier = 0    

fig, ax = plt.subplots(layout='constrained')

for attribute, measurement in scores.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
# ax.set_ylabel(metric_name)
ax.set_xlabel("Datasets")
ax.set_title(f"Resemblance scores of samples with and without watermark")
ax.set_xticks(x + width, model_names)
ax.legend(loc='upper left', ncols=3)
ax.set_ylim(0, 1.1)
plt.savefig(f"alobetween_datasets.pdf")
plt.show()
```
