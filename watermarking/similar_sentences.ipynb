```python
import nltk
from nltk.corpus import words
import random

# nltk.download('words')

vocab = words.words()

vocab = random.sample(vocab, 10000)
```

```python
from sentence_transformers import SentenceTransformer, util

# Load pre-trained model
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')

def get_embedding(text):
    return model.encode(text, convert_to_tensor=True)

def partition_vocabulary(vocabulary, seed_phrase):
    # Encode the seed phrase
    seed_embedding = get_embedding(seed_phrase)
    # print(seed_embedding)
    
    # Encode vocabulary words
    vocab_embeddings = {word: get_embedding(word) for word in vocabulary}
    print(vocab_embeddings)
    
    # Calculate similarities
    similarities = {word: util.pytorch_cos_sim(seed_embedding, embed).item() for word, embed in vocab_embeddings.items()}
    print(similarities)
    
    sorted_similarity_scores = sorted(similarities.items(), key=lambda x: x[1], reverse=True)
    
    sorted_similarity_dict = {k: v for k, v in sorted_similarity_scores}
    
    return sorted_similarity_dict

def count_common_words(sorted_vocab1, sorted_vocab2, percent):
    top_n = int(len(sorted_vocab1) * percent)
    top_vocab1 = set(word for _, word, _ in sorted_vocab1[:top_n])
    top_vocab2 = set(word for _, word, _ in sorted_vocab2[:top_n])
    common_words = top_vocab1.intersection(top_vocab2)
    return len(common_words), common_words


seed_phrase1 = "education number"
seed_phrase2 = "num edu"

sorted_vocab_1 = partition_vocabulary(vocab, seed_phrase1)
# sorted_vocab_2 = partition_vocabulary(vocab, seed_phrase2)
# 
# common_word_count, common_words = count_common_words(sorted_vocab_1, sorted_vocab_2, 0.25)
# 
# print(common_word_count / len(sorted_vocab_1))
```
